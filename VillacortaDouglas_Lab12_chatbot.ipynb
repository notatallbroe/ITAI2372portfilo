### Introduction:
 1. What is a chatbot? A chatbot (a robot that chats with human beings) is an application that generates a conversation with a user.
 2. Types of chatbots:
        - Button based: The most basic chatbot that interacts with the user by cliking on buttons.
        - Rule-based: A simple chatbot that interacts with logical rules of if-then. It is based on simple FAQ (Frequently Asked Questions).
        - AI-powered: It is a type of rule-based bot that is backed by AI predefined options for questions and answers. This type is equipped with AI and NLU (Natural Language Understanding).
        - Generatice AI: These types of bots are relatively advanced and they are based on LLM (Large Language Models) with high-quality texts, images, and sound.
        - Retrieval-based Chatbot: This type is the opposite of generative AI-based chatbot. Data is not generated by AI instantaneously; but has already been created and it will only be retrieved by the chatbot in response to the user's question
        - Voice Enabled Chatbot: This type of chatbot runs a conversation with a user by speaking to the user rather than typing.
        - Hybrid: A hybrid chatbot combines rule-based, ML models and logic; and integration of AI technologies.
